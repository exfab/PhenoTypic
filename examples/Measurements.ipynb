```python
from phenotypic.data import load_colony, make_synthetic_colony
from phenotypic.detect import OtsuDetector
from phenotypic.objedit import SmallObjectRemover
from phenotypic import Image, ImagePipeline
import phenotypic.measure

import importlib
import inspect
import phenotypic.measure

# Dynamically get the module object
measure_module = importlib.import_module("phenotypic.measure")

# Load all classes defined in the module
all_classes = [
    cls for name, cls in inspect.getmembers(measure_module, inspect.isclass)
    if cls.__module__.startswith(measure_module.__name__)
]
pipeline = ImagePipeline(
        ops=[
            OtsuDetector(),
            SmallObjectRemover(min_size=10)
        ],
        meas=[cls() for cls in all_classes])

image = Image(make_synthetic_colony(h=25, w=25))
image.show()
```

```python

image = pipeline.apply(image)
image.show_overlay()
```
