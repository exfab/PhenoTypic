```python
import numpy as np

import phenotypic as pty

rgb_array = pty.data.make_synthetic_colony(h=50, w=50, seed=0)

colony_image = pty.Image(rgb_array, imformat='RGB')
colony_image.show()
```

```python
from phenotypic.detection import OtsuDetector

colony_image = OtsuDetector().apply(colony_image)
colony_image.show_overlay()
```

```python
from phenotypic.enhance import GaussianBlur

colony_image = GaussianBlur(sigma=5).apply(colony_image)
colony_image = OtsuDetector().apply(colony_image)
colony_image.show_overlay()
```

```python
np.savez(f'../src/phenotypic/data/synthetic_colony.npz', array=colony_image.array[:], objmask=colony_image.objmask[:])
```
